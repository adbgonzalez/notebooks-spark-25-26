```python
# Inicializamos SparkSession y SparkContext
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("rdd-test") \
    .getOrCreate()

sc = spark.sparkContext
print(spark.version)  # Verifica la versión de Spark
```

```python
rdd1 = sc.parallelize(range(10))
```

```python
rdd1.collect()
```

```python
rdd2 = sc.parallelize( "Esta é unha proba para facer probas, polo menos unha".split())
```

```python
rdd2.collect()
```

```python
rdd_pair = rdd2.map(lambda a: (a,1))
rdd_pair.collect()
```

```python
rdd_pair.countByKey()
```

```python
rdd_q = sc.textFile('/data/el_quijote.txt')
rdd_q.take(3)
```

```python
rdd_palabras = rdd_q.flatMap(lambda a: a.split())
```

```python
rdd_palabras.take(100)
```

```python
rdd_pares = rdd_palabras.map(lambda a : (a,1))
rdd_pares.take(20)
```

```python
wordcount = rdd_pares.countByKey()
#wordcount
```

```python
rdd_iniciales = rdd_palabras.keyBy(lambda p: p[0])
rdd_iniciales.take(10)
```

```python
rdd_longitudes = rdd_palabras.keyBy(lambda p: len(p))

rdd_longitudes.take(10)
```

```python
rdd_iniciales.map(lambda x: (x[0].lower(),x[1].lower())).filter(lambda x: x[0] == 'a').values().count()
```

```python
keys = rdd_iniciales.keys().distinct().collect()
fractions = {k: 0.0 for k in keys}   # por defecto 0
fractions['a'] = 0.8
fractions['b'] = 0.2
rdd_iniciales.sampleByKey(withReplacement=False, fractions=fractions,seed=42).collect()
```

```python
rdd_iniciales.take(5)
```

```python
cosa = rdd_iniciales.groupByKey()
```

```python
for (key,values) in cosa.collect():
    lista = list(values)
    print("valores de ",key,": ",lista)
```

```python
rdd_iniciales.reduceByKey(lambda x,y: x if len(x) < len(y) else y).sortByKey().collect()
```
